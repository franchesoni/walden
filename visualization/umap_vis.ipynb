```python
import umap
reducer = umap.UMAP()
```

```python
import numpy as np
Xl = np.load("X_labeled.npy")
y = np.load("y_labeled.npy")
print(Xl.shape)
two_dimensional_Xl = reducer.fit_transform(Xl)
two_dimensional_Xl_no_other = reducer.fit_transform(Xl[y!=3])
```

```python
y = np.load("y_labeled.npy") 
print(y.shape, y.min(), y.max())
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(two_dimensional_Xl[:, 0], two_dimensional_Xl[:, 1])
plt.figure()
plt.scatter(two_dimensional_Xl[:, 0], two_dimensional_Xl[:, 1], c=y)

plt.figure()
plt.scatter(two_dimensional_Xl_no_other[:, 0], two_dimensional_Xl_no_other[:, 1])
plt.figure()
plt.scatter(two_dimensional_Xl_no_other[:, 0], two_dimensional_Xl_no_other[:, 1], c=y[y!=3])

```
